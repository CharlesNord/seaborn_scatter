```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
x = np.random.normal(0, 1, (100,))
y = np.random.normal(0, 1, (100,))
label = np.random.randint(0, 10, 100)
```

```python
palette = ListedColormap(sns.color_palette('hls',10))
```

```python
plt.scatter(x, y, c=label, cmap=palette)
plt.colorbar()
```

```python
def gaussian_mixture(batchsize, label_indices, num_labels):
    def sample(x, y, label, num_labels):
        shift = 1.4
        r = 2.0*np.pi/float(num_labels)*label
        new_x = x*np.cos(r)-y*np.sin(r)
        new_y = x*np.sin(r)+y*np.cos(r)
        new_x += shift*np.cos(r)
        new_y += shift*np.sin(r)
        return np.array([new_x, new_y]).reshape((2,))
    
    x_var = 0.5
    y_var = 0.05
    x = np.random.normal(0, x_var, batchsize)
    y = np.random.normal(0, y_var, batchsize)
    z = np.empty((batchsize, 2), dtype=np.float32)
    for batch in range(batchsize):
        z[batch, 0:2] = sample(x[batch], y[batch], label_indices[batch], num_labels)
    return z
```

```python
batchsize = 1000
label = np.ones((100, 1))
label_indices = []
for i in range(10):
    label_indices.append(label*i)
    
label_indices = np.array(label_indices)
label_indices = label_indices.reshape(1000,)
z = gaussian_mixture(batchsize, label_indices, 10)
```

```python
plt.scatter(z[:,0],z[:,1],c=label_indices, cmap=palette, s=3)
plt.colorbar()
```

```python
def swiss_roll(batchsize, label_indices, num_labels):
    def sample(label, num_labels):
        uni = np.random.uniform(0.0, 1.0)/num_labels+float(label)/float(num_labels)
        r = np.sqrt(uni)*3.0
        rad = np.pi*4.0*np.sqrt(uni)
        x = r*np.cos(rad)
        y = r*np.sin(rad)
        return np.array([x, y]).reshape((2,))
    
    z = np.zeros((batchsize, 2), dtype=np.float32)
    for batch in range(batchsize):
        z[batch, 0:2] = sample(label_indices[batch], num_labels)
    
    return z
```

```python
batchsize = 1000
label = np.ones((100, 1))
label_indices = []
for i in range(10):
    label_indices.append(label*i)
    
label_indices = np.array(label_indices)
label_indices = label_indices.reshape(1000,)
z = swiss_roll(batchsize, label_indices, 10)
```

```python
plt.scatter(z[:,0],z[:,1],c=label_indices, cmap=palette, s=3)
plt.colorbar()
```
